Você trabalha como analista de dados em uma loja de roupas que deseja entender as tendências de vendas de um determinado item. A equipe de planejamento de estoque precisa de uma análise clara das tendências para identificar períodos de alta e baixa demanda. Uma das abordagens sugeridas é calcular as médias móveis, o que ajudará a suavizar as flutuações nos dados e a identificar padrões de vendas.

Objetivo: Calcular e visualizar médias móveis simples (para 7 e 30 dias) para a série temporal de vendas e interpretar os resultados para prever o comportamento de vendas.

Insira o código abaixo de cada comentário.


```python
# Importando bibliotecas necessárias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.ar_model import AutoReg

# Gerando dados simulados de vendas diárias
np.random.seed(42)
dias = pd.date_range(start="2023-01-01", periods=180, freq="D")
vendas = 50 + 5 * np.sin(np.linspace(0, 6 * np.pi, 180)) + np.random.normal(0, 2, 180)
serie_vendas = pd.Series(vendas, index=dias)

# Visualizando os dados históricos de vendas]

np.random.seed(42)


model = AutoReg(vendas, lags=2)
model_fit = model.fit()


forecast_steps = 5
predictions = model_fit.predict(start=len(vendas), end=len(vendas) + forecast_steps - 1)



plt.figure(figsize=(10, 5))
plt.plot(vendas, label='Série Original')
plt.plot(range(len(vendas), len(vendas) + forecast_steps), predictions, label='Previsão', color='red', marker='o')
plt.title('Modelo Autoregressivo Simples')
plt.xlabel('Tempo')
plt.ylabel('Valor')
plt.legend()
plt.grid(True)
plt.show()


# Cálculo de médias móveis simples 7


np.random.seed(42)


vendas = pd.Series(np.cumsum(np.random.randn(100) + 2), index=pd.date_range(start='2020-01-01', periods=100))

window_size = 7
moving_average = vendas.rolling(window=window_size).mean()

# Plotando a série original e a média móvel
plt.figure(figsize=(12, 6))
plt.plot(vendas, label='Série Original')
plt.plot(moving_average, label=f'Média Móvel ({window_size} dias)', color='orange')
plt.title('Exemplo de Média Móvel em Série Temporal')
plt.xlabel('vendas')
plt.ylabel('Valores')
plt.legend()
plt.grid(True)

# Cálculo de médias móveis simples 30


np.random.seed(42)


vendas = pd.Series(np.cumsum(np.random.randn(100) + 2), index=pd.date_range(start='2020-01-01', periods=100))

window_size = 30
moving_average = vendas.rolling(window=window_size).mean()

# Plotando a série original e a média móvel
plt.figure(figsize=(12, 6))
plt.plot(vendas, label='Série Original')
plt.plot(moving_average, label=f'Média Móvel ({window_size} dias)', color='orange')
plt.title('Exemplo de Média Móvel em Série Temporal')
plt.xlabel('vendas')
plt.ylabel('Valores')
plt.legend()
plt.grid(True)

```
